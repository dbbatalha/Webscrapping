```python
import requests
import urllib.request
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup,NavigableString
```

```python
#  THIS IS THE SCRIPT FOR SCRAPE THE DATABASE FROM IBJJF FROM BRASILEIRO BJJ COMPETITION

data = []
categories = []
a=[]
listz=[]


for i in range(1996,2014):      # Number of pages plus one 
    if i<= 2012:
        url='https://ibjjf.com/events/results/{}-campeonato-brasileiro-de-jiu-jitsu'.format(i)
        response = requests.get(url)
        soup = BeautifulSoup(response.text,"html.parser")
        result = soup.find('div', class_ = 'col-sm-12 col-md-6 athletes-result')



        for new in result.find_all('div', {'class':'row results'},'tr'):
            for athletes in new.find_all('tr'):
                data.append(athletes.find('td',class_ = 'place').text)
                data.append(athletes.find('div',class_ = 'athlete-name').text)
                data.append(athletes.find('div',class_ = 'academy-name').text)
                data.append(soup.find('h3', {'class':'title-section text-center'}).text)

        for new in result.find_all('div',{'class':'category mt-4 mb-3'}):
            categories.append(new.get_text(strip=True))
    else:
        brasileiro = [107, 172, 258, 380, 515, 712, 894,1176]
        for j in brasileiro:
            url='https://www.ibjjfdb.com/ChampionshipResults/{}/PublicResults'.format(j)
            response = requests.get(url)
            soup = BeautifulSoup(response.text,"html.parser")
            result = soup.find('div', class_ = 'col-xs-12 col-md-6 col-athlete')

            for new in result.find_all('div', {'class':'list'}):
                for athletes in new.find_all('div', {'class':'athlete-item'}):
                    data.append(athletes.find('div',{"class": 'position-athlete'}).get_text(strip=True).replace("\r""\n"'                         ', ""))
                    data.append(athletes.find('p').get_text(strip=True).replace("\r""\n"'                         ', ""))
                    data.append(athletes.find('span').get_text(strip=True).replace("\r""\n"'                         ', ""))
                    data.append(soup.find_all('h2', {'class':'title'})[1].text.replace("\r\n"" ", ""))

            for new in result.find_all('h4',{'class':'subtitle'}):
                categories.append(new.get_text(strip=True).replace("\r""\n"'                         ', ""))
                

inner_size = 4
finaldata = [ data[i:i+inner_size] for i in range(0, len(data), inner_size) ]
dataf = pd.DataFrame(finaldata)
dataf.columns =['Place','Athlete','Team','Tournament']




# copying the colum PLACE of the dataf and excluding the first line of this one
newplace= dataf[['Place']].copy()
newplace2= newplace.iloc[1:]
# transform in a list
newplace3= dataf['Place'].tolist()

#transform places of athletes in 1 or -1 and deleting the first row
a = [-1 if i == '1' else 1 for i in newplace3]
del a[0]

index = 0
while len(listz)<len(a):
    for y in a:
        if y>=0:
            listz.append(categories[index])
        else:
            index = (index + 1) % len(categories)
            listz.append(categories[index])


#matching both dataframes
listz.insert(0, categories[0])

datafz = pd.DataFrame(listz)
datafz.columns =['Category']

d=pd.concat([dataf, datafz], axis=1)
d
```
